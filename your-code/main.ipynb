## **Exploratory Data Analysis**

### What's the genre with the highest median popularity?

```python
df_genres = df.dropna(subset=["genres"]).copy()
df['genres'].unique()
```

```python
df_genres["genres"] = df_genres["genres"].apply(lambda x: [genre.strip() for genre in x.split("|")])

df_exploded = df_genres.explode("genres")
df_exploded
```

```python
median_popularity = df_exploded.groupby("genres")["popularity"].median().sort_values(ascending=False)
highest_genre = median_popularity.idxmax()
highest_value = median_popularity.max()
display(median_popularity.head(3))
print(f"The genre with the highest median popularity is '{highest_genre}' with a median popularity of {highest_value:.2f}.")
```

### What's the genre with the highest median revenue?

```python
median_revenue = df_exploded.groupby("genres")["revenue"].median().sort_values(ascending=False)
highest_genre_revenue = median_revenue.idxmax()
highest_value_revenue = median_revenue.max()
display(median_revenue.head(3))
print(f"The genre with the highest median revenue is '{highest_genre_revenue}' with a median revenue of {highest_value_revenue:.2f}.")
```

### What's the genre with the highest median vote_count?

```python
median_vote_count = df_exploded.groupby("genres")["vote_count"].median().sort_values(ascending=False)
highest_genre_vote_count = median_vote_count.idxmax()
highest_value_vote_count = median_vote_count.max()
display(median_vote_count.head(3))
print(f"The genre with the highest median vote count is '{highest_genre_vote_count}' with a median vote count of {highest_value_vote_count:.2f}.")
```

### What's the genre with the highest mean vote_average?

```python
mean_vote_average = df_exploded.groupby("genres")["vote_average"].mean().sort_values(ascending=False)
highest_genre_vote_average = mean_vote_average.idxmax()
highest_value_vote_average = mean_vote_average.max()
display(mean_vote_average.head(3))
print(f"The genre with the highest mean vote average is '{highest_genre_vote_average}' with a mean vote average of {highest_value_vote_average:.2f}.")
```

### What movie is considered movie genre is considered the "best"?


```python
df_exploded.sort_values(by="vote_average", ascending=False).drop_duplicates(subset=["genres"], keep="first")[["original_title", "genres", "vote_average"]]
```

### What month is considered "best" for releasing a films/shows?


```python

df['release_date'] = pd.to_datetime(df['release_date'], format='%m/%d/%y', errors='coerce')
df['release_month'] = df['release_date'].dt.month
month_mapping = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
df['release_month_name'] = df['release_month'].map(month_mapping)

print(df[['release_month', 'release_month_name']].head())

```

```python

mean_revenue_by_month = df.groupby('release_month_name')['revenue'].mean().sort_values(ascending=False)
highest_month_revenue = mean_revenue_by_month.idxmax()
highest_value_month_revenue = mean_revenue_by_month.max()
display(mean_revenue_by_month)
print(f"The month with the highest mean revenue is '{highest_month_revenue}' with a mean revenue of {highest_value_month_revenue:.2f}.")
```

# Conclusions

1. The Adventure genre stands out on several key metrics—it has the highest median revenue, vote count, and popularity. This suggests that films in the Adventure genre consistently attract audience interest and generate substantial box office income.

2. The Documentary genre exhibits the highest mean vote average. Although its absolute revenue and popularity numbers might be lower than blockbuster figures, the high average rating indicates that documentaries are highly appreciated for their quality.

3. The differences between mean and median metrics across genres indicate heterogeneous distributions, potentially due to outliers or skewed data.

# Limitations
1. Genre Handling: Films belonging to multiple genres were "exploded" into separate rows. As a result, the same movie may appear in multiple genre categories, which can affect the aggregated statistics for each genre.
2. Statistical Aggregation: Relying on mean and median values offers a general overview but does not capture the full distribution of the metrics. The presence of outliers or skewed data might cause these aggregated metrics to be misleading.
